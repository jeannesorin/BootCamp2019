```python
### PROBLEM 3
import numpy as np
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt


baseball = np.load("MLB.npy")


fig, ax = plt.subplots(1,3)

#plt.plot(x, y, '.')
#plt.plot(x, b + m * x, '-')
# Correlation height / weight
b, m = polyfit(baseball[:,0], baseball[:,1], 1)
ax[0].set(xlabel='Weight', ylabel='Height')
ax[0].plot(baseball[:,0], baseball[:,1], "m.")
ax[0].plot(baseball[:,0], b + m * baseball[:,0], 'm-')


# Correlation height / age
b, m = polyfit(baseball[:,0], baseball[:,2], 1)
ax[1].set(xlabel='Weight', ylabel='Age')
ax[1].plot(baseball[:,0], baseball[:,2], "g.")
ax[1].plot(baseball[:,0], b + m * baseball[:,0], 'g-')


# Correlation weight / age
b, m = polyfit(baseball[:,1], baseball[:,2], 1)
ax[2].set(xlabel='Height', ylabel='Age')
ax[2].plot(baseball[:,1], baseball[:,2], "b.")
ax[2].plot(baseball[:,1], b + m * baseball[:,1], 'b-')

plt.suptitle("Baseball players' characteristics (correlations)")

plt.tight_layout()
plt.show()

# Could be improved
```

```python
### PROBLEM 5

x = np.linspace(-2.0, 2.0,1000)
X, Y = np.meshgrid(x, x)
Z = (1 - X)**2 + (100)*((Y - X**2)**2)

fig, ax = plt.subplots(1,2)
ax[0].pcolormesh(X,Y,Z, cmap="inferno")
ax[0].plot(1,1, "yx")
cas = ax[0].pcolormesh(X,Y,Z, cmap="inferno")
ax[1].contour(X,Y,Z, cmap="viridis")
ax[1].plot(1,1, "yx")
das = ax[1].contour(X,Y,Z, cmap="viridis")

fig.colorbar(cas, ax=ax[0])
fig.colorbar(das, ax=ax[1])

plt.tight_layout()


plt.show()
```
