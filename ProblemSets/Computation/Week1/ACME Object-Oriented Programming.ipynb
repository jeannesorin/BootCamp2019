```python
### PROBLEM 1

class Backpack:
    """A Backpack object class. Has a name, a maximum capacity, a color and a list of contents.
    Attributes:
        name (str): the name of the backpack's owner.
        color (str): the color of the backpack
        max_size (int): the maximum number of items in the backpack
        contents (list): the contents of the backpack.  
    Methods:
        put(): add an item in the backpack
        dump(): empty the backpack
    """
    def __init__(self, name, color, max_size = 5):
        self.name = name
        self.color = color
        self.max_size = max_size
        self.contents = []
        
    def put(self, item):
        if len(self.contents) >= self.max_size:
            print("No Room!")
        if len(self.contents) < self.max_size:
            self.contents.append(item)
            
    def dump(self):
        self.contents = []
        
    
def test_backpack():
    testpack = Backpack("Barry", "black")       # Instantiate the object.
    if testpack.name != "Barry":                # Test an attribute.
        print("Backpack.name assigned incorrectly")
    for item in ["pencil", "pen", "paper", "computer"]:
        testpack.put(item)                      # Test a method.
    print("Contents:", testpack.contents)

test_backpack()        
        
```

```python
### PROBLEM 4

class ComplexNumber:
    """A Complex Number class with a real and an imaginary part
    Attributes:
        real (int): the real component of the complex number
        imag (int): the imaginary component of the complex number
    """
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
        
    def conjugate(self):
        return(ComplexNumber(self.real, -1*self.imag))
    
    def __str__(self):
        if self.imag < 0:
            mylist = ["(", self.real, "-", -1*self.imag, "j)"]
            String = "".join(map(str, mylist))
            return(String)
        if self.imag >= 0:
            mylist = ["(", self.real, "+", self.imag, "j)"]
            String = "".join(map(str, mylist))
            return(String)
        
    def __abs__(self):
        return((self.real**2 + self.imag**2)**0.5)
    
    def __eq__(self, other):
        if (self.real == other.real) & (self.imag == other.imag):
            return("True")
        else:
            return("False")
        
    def __add__(self, other):
        self.real = self.real + other.real
        self.imag = self.imag + other.imag
        
    def __sub__(self, other):
        self.real = self.real - other.real
        self.imag = self.imag - other.imag

    def __mul__(self, other):
        self.real = self.real * other.real
        self.imag = self.imag * other.imag
 
    def __truediv__(self, other):
        self.real = self.real / other.real
        self.imag = self.imag / other.imag

```
