```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
### PROBLEM 1

def function_pb1(n):
    ar = np.random.normal(size=(n,n))
    mean = ar.mean(axis = 1)
    return(mean.var())

def plot_pb1():
    result = []
    
    for numb in np.arange(100,1000, 100):
        result.extend([function_pb1(numb)])
    
    plt.plot(np.arange(100,1000, 100), result)
    
    
plot_pb1()
    
```

```python
### PROBLEM 2

def function_pb2():
    
    # Initialization
    cosx = []
    sinx = []
    
    for numb in np.linspace(-2*np.pi,2*np.pi,100):
        cosx.extend([np.cos(numb)])
        sinx.extend([np.sin(numb)])

    plt.plot(
        np.linspace(-2*np.pi,2*np.pi,100), cosx,
        np.linspace(-2*np.pi,2*np.pi,100), sinx)
    
function_pb2()
```

```python
### PROBLEM 3

def function_pb3():
    x1 = np.linspace(-2,0.999, 100)
    x2 = np.linspace(1.001, 6, 300)
    y1 = 1 / (x1-1)
    y2 = 1 / (x2-1)
    #print(y1)
    #print(y2)
    
    plt.plot(x1, y1,'m--',
             x2, y2, 'm--', linewidth=4)#plt.plot(x, y)
    plt.ylim(-6,6)
    plt.xlim(-2, 6)
    
function_pb3()
```

```python
### PROBLEM 4


def function_pb4():

    # Initialization
    x = np.linspace(0,2*np.pi,100)
    sinx = np.sin(x)
    sin2x = np.sin(2*x)
    s2inx = 2*np.sin(x)
    s2in2x = 2*np.sin(2*x)
    
    fig, ax = plt.subplots(2,2,sharex=True, sharey=True)

    ax[0,0].set_ylim([-2,2])
    ax[0,0].plot(x, sinx, 'g-')
    ax[0,0].set_title(r"$sin(x)$")
    ax[0,1].plot(x, sin2x, 'r--')
    ax[0,1].set_title("sin(2x)")
    ax[1,0].set_ylim([-2,2])
    ax[1,0].plot(x, s2inx, 'b--')
    ax[1,0].set_title("2sin(x)")
    ax[1,1].plot(x, s2in2x, 'm:')
    ax[1,1].set_title("2sin(2x)")
    plt.suptitle("Trigonometry")
    plt.tight_layout()

function_pb4()
```

```python
### PROBLEM 5

def function_pb5():
    
    # Load dataset
    data = np.load("FARS.npy")
    hours = data[:,0]
    long = data[:,1]
    lat = data[:,2]
    fig, ax = plt.subplots(1,2)

    #ax[0,0].set_ylim([-2,2])
    ax[0].plot(long, lat, "o", markersize = 0.5)
    ax[0].set_aspect("equal")
    ax[0].set_title("Location of fatalities", fontsize=10)

    #ax2 = plt.subplot(1,2,2)
    ax[1].hist(hours, bins = np.arange(0,24))
    ax[1].set_title("Time histogram of fatalities", fontsize=10)

    
    plt.suptitle("Fatalities in the US")
    plt.tight_layout()
    #plt.show()
    
function_pb5()
```
